```python
import pandas as pd
```

```python
data = {
    "Name": ["Ram", "Sita", "John", "Asha"],
    "Maths": [85, 90, 78, 92],
    "Science": [88, 91, 80, 95],
    "English": [84, 89, 79, 90]
}

df = pd.DataFrame(data)

print(df)
```

```python
df["Average"] = (df["Maths"] + df["Science"] + df["English"]) / 3

print(df)
```

```python
top_student = df[df["Average"] == df["Average"].max()]

print("Top Student:")
print(top_student)
```

```python
df.to_csv("student_data.csv", index=False)

print("CSV File Saved")
```
